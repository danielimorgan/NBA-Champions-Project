# players_stats_by_season_full_details.CSV Modification

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# Load file
players_data_loading = Path("players_stats_by_season_full_details.csv")

# Read file
players_data = pd.read_csv(players_data_loading)

# Show file

players_data.head(2)
```

```python
# Drop initial columns not needed.
players_data = players_data.drop(
    columns=['birth_date', 'height', 'height_cm', 'weight',
    'weight_kg', 'nationality', 'high_school', 'draft_round',
    'birth_month', 'birth_year', 'draft_pick', 'draft_team'])

players_data
```

```python
# Drop all leagues except the NBA. 
nba_only = players_data.query(
    "League == ['NBA']")

nba_only
```

```python
# Fix the Season column.
season_split = nba_only['Season'].str.split(
    pat=" ",
    expand=True)

season_split['Season'] = season_split[0] + "-" + season_split[2]

season_split = season_split.drop(columns=[0,1,2])
season_split
```

```python
# Merge the new season data to the original dataframe.
nba_only = pd.merge(
    nba_only,
    season_split,
    left_index=True,
    right_index=True)

nba_only
```

```python
# Drop duplicate Season and rename the other.
nba_only = nba_only.drop(columns='Season_x')

nba_only = nba_only.rename(columns={'Season_y':'Season'})

nba_only
```

```python
# Create the new Dataframe and reindex.
player_total_stats_complete = pd.DataFrame(data=nba_only)

player_total_stats_complete = player_total_stats_complete.reset_index(drop=True)

player_total_stats_complete
```

```python
# Save and transport new CSV.
player_total_stats_complete.to_csv("Data Mods/Player Total Stats By Season.csv")
```
